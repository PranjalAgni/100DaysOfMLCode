```
i = 0
mod_age =  []
mod_diab = []
while i < 767:
  row_data = my_data.iloc[i]
  age = int(row_data['Age (years)'])
  sum_diab = 0
  count=0
  changing_age = age
  while age == changing_age:
        i += 1
        if i == 767:
           break
        changing_row = my_data.iloc[i]
        changing_age = int(changing_row['Age (years)'])
        sum_diab += int(changing_row['Diabetic'])
        count += 1
  print('Number of person with age {} are {}'.format(age,count))
  mod_age.append(age)
  if count == 0:
    mod_diab.append(0)
  else:
    mod_diab.append(sum_diab/count)
  
  
      
```

# Lets Apply some machine learning

**Starting with Linear Regression**

```
from statistics import mean
```

```
# Get our sanatized data
diab_data = np.array(mod_diab , dtype=np.float64)
person_age = np.array(mod_age , dtype=np.float64)
```

```
# Calculating slope of the line
# Calculating Y Intercept
# b = mean(y) - m*mean(x)
def line_best_fit(age_x, diab_y):
    res =  (mean(age_x) * mean(diab_y) - mean(age_x * diab_y)) / ((mean(age_x)**2) - (mean(age_x**2)))
    b = mean(diab_y) - res*mean(age_x)
    return res,b
```

```
m,b = line_best_fit(person_age , diab_data)
```

# Lets Start Predicting

```
# what is the chances of diabetes if age of person is 22
age = 60
pred_res = m*age + b
pred_res
```

```
regression_line = [(m*x+b) for x in person_age]
```

```
plt.scatter(person_age,diab_data)
plt.scatter(age , pred_res, color='g')
plt.plot(person_age,regression_line)
plt.show()
```

**Testing how best fit our line is**

```
def squarred_error(y_org , y_pred):
  return sum((y_pred-y_org)**2)
```

```
def coff_of_determination(y_data , y_regression):
  y_mean = [mean(y_data) for y in y_data]
  sqrd_err_reg = squarred_error(y_data , y_regression) 
  sqrd_err_mean = squarred_error(y_data, y_mean)
  return (1- (sqrd_err_reg/sqrd_err_mean))
```

```
r_sqrd = coff_of_determination(diab_data , regression_line)
r_sqrd
```

*Linear regression does not fit our curve , too good its rsquarred value is less , according to our graph we should try polynomial regression  *
