# What you should know to get a job at Amazon?

**We have a dataset of developers got selected in Amazon with their basic qualification**

**Dataset is picked from Kaggle you can [check it out](https://www.kaggle.com/atahmasb/amazon-job-skills).Also all the data is about those developers who got selected**

***Lets try to generate a wordcloud which was in common in most of developers Resume***

```
# import dependencies
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Collect the data
df = pd.read_csv('amazon_jobs_dataset.csv')
df.shape
```

```
df.head()
```

```
# Let's check for null values in the dataset
df.isnull().sum()
```

**Let's know about the dataset**

1.   **DESCRIPTION: ** *About the job profile,on what field you will be working on*
2.   **BASIC QUALIFICATIONS: ** *Experience of candidates,their skills*
3.   **PREFERRED QUALIFICATIONS: ** *Required skills for the job*



```
# Lets remove the null values
df = df.dropna(subset=['BASIC QUALIFICATIONS', 'PREFERRED QUALIFICATIONS'])
```

```
# Let's build our wordcloud of the qualification of developers

from wordcloud import WordCloud

wordcloud_basic = WordCloud(
    background_color = 'black',
    max_words =  200,
    max_font_size =  40,
    scale=3,
    random_state=42
).generate(str(df['BASIC QUALIFICATIONS']))

fig = plt.figure(1, figsize=(15,30))
plt.axis('off')
plt.imshow(wordcloud_basic)
plt.show()
```

```
# Lets build a wordcloud for the required qualifications
from wordcloud import WordCloud
wordcloud_prefer = WordCloud(
    background_color='black',
    max_words=200,
    max_font_size=40,
    scale=3,
    random_state=42
).generate(str(df['PREFERRED QUALIFICATIONS']))


fig = plt.figure(1, figsize=(15,30))
plt.axis('off')
plt.imshow(wordcloud_prefer)
plt.show()
```
